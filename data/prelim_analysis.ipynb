```python
df_res = pd.DataFrame()
for ticker in list_ticker_snp_500:
    df_tmp = si.get_data(ticker,start_date='09-11-2020')['close'].diff()
    df_res[ticker] = df_tmp
```

```python
df_res
```

```python
df_res.drop(index=df_res.index[0],inplace=True)
```

```python
df_res
```

```python
corrs = df_res.corr()
corrs
```

```python
df1 = corrs.abs().unstack().sort_values(kind='quicksort',ascending=False).drop_duplicates()
```

```python
df1.to_frame().reset_index()
```

```python
df1.drop(df1.index[:4], inplace=True)
```

```python
df1 = df1.to_frame().reset_index()
```

```python
df1
```

```python
merged1 = pd.merge(df1,snp_500[['GICS Sector', 'Symbol']], how='left', left_on='level_0', right_on='Symbol')
```

```python
merged1
```

```python
merged2 = pd.merge(merged1,snp_500[['GICS Sector', 'Symbol']], how='left', left_on='level_1', right_on='Symbol')
```

```python
merged2
```

```python
merged2.columns = ['level_0','level_1','corrs', 'GICS Sector_x', 'Symbol_x', 'GICS Sector_y', 'Symbol_y']
```

```python
merged2[['GICS Sector_x', 'GICS Sector_y']].value_counts(normalize=True)
```

```python
merged2.groupby(['GICS Sector_x','GICS Sector_y'])['corrs'].mean().sort_values(ascending=False).head(20)
```

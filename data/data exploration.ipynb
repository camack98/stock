## Analysis of some data

For this part I will just explore a `AAPL` (Apple)

```python
si.get_company_info('AAPL')
```

```python
df_AAPL = si.get_data('AAPL')
```

```python
df_AAPL
```

```python
%matplotlib notebook
df_AAPL[['open','close']].plot()
```

```python
df_AAPL[['open','close']].plot(logy=True)
```

```python
df_AAPL.loc[df_AAPL.index >pd.to_datetime('11-09-2020'), ['open','close']].plot()
```

```python
df_AAPL['diff_close_open'] = df_AAPL['close'] - df_AAPL['open']
```

```python
df_AAPL.loc[df_AAPL.index >pd.to_datetime('11-09-2020'), ['diff_close_open']].plot()
```

```python
small_AAPL = df_AAPL[df_AAPL.index >pd.to_datetime('11-09-2020')]
```

```python
%matplotlib notebook
plt.plot(small_AAPL['diff_close_open'], small_AAPL['volume'],'.')
plt.xlabel('diff_close_open')
plt.ylabel('volume')
plt.show()
```

```python
small_AAPL['close_lag1'] = small_AAPL.loc[:,'close'].shift()
small_AAPL['diff_close'] = small_AAPL.loc[:,'close'].diff()
```

```python
small_AAPL
```

```python
small_AAPL['diff_close'].plot()
```

```python
import seaborn as sns
%matplotlib notebook

corr = small_AAPL.corr()
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
plt.tight_layout()
```

# Now looking at some autocorrelation

```python
%matplotlib notebook
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plot_acf( small_AAPL['close'],lags=150)
```

```python
plot_acf( small_AAPL['diff_close'].dropna(),lags=50)
```

# Let's analyse some ticker together

```python
import random
random.seed(42)
sub_list = random.sample(list_ticker_snp_500,20)
sub_list
```

```python
df_res = pd.DataFrame()
for ticker in sub_list:
    df_tmp = si.get_data(ticker,start_date='09-11-2020')
    df_res[ticker] = df_tmp['close']
```

```python
df_res
```

```python
%matplotlib notebook
corr = df_res.corr()
sns.clustermap(corr)
plt.tight_layout()
```
